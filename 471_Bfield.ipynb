```python
import pandas as pd
import plotly.express as px
```

```python
url1 = 'https://raw.githubusercontent.com/hedbergj/CCNY-PHYS471/main/experimental-data/bfield-unknownN/bfield-solenoid-1.csv'
df1 = pd.read_csv(url1)

url2 = 'https://raw.githubusercontent.com/hedbergj/CCNY-PHYS471/main/experimental-data/bfield-unknownN/bfield-solenoid-2.csv'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/hedbergj/CCNY-PHYS471/main/experimental-data/bfield-unknownN/bfield-solenoid-3.csv'
df3 = pd.read_csv(url3)

url4 = 'https://raw.githubusercontent.com/hedbergj/CCNY-PHYS471/main/experimental-data/bfield-unknownN/bfield-solenoid-4.csv'
df4 = pd.read_csv(url4)
```

Code used for plotting multible data on the same figure is from community.plotly.com *How to plot multiple lines on the same y-axis using plotly express?* access by the link https://community.plotly.com/t/how-to-plot-multiple-lines-on-the-same-y-axis-using-plotly-express/29219/18

```python
import plotly.graph_objects as go
fig = go.Figure([
#fig.add_trace(
    go.Scatter(
        name='Measurement 1',
        x=df1['relative_time [ms]'],
        y=df1['Bfield [microteslas]'],
        #mode='markers+lines',
         mode='markers',
        marker=dict(color='red', size=2),
        showlegend=True
    ),
     go.Scatter(
        name='Measurement 2',
        x=df2['relative_time [ms]'],
        y=df2['Bfield [microteslas]'],
        mode='markers',
        marker=dict(color='blue', size=2),
        showlegend=True
    ),
         go.Scatter(
        name='Measurement 3',
        x=df3['relative_time [ms]'],
        y=df3['Bfield [microteslas]'],
        mode='markers',
        marker=dict(color='black', size=2),
        showlegend=True
    ),
         go.Scatter(
        name='Measurement 4',
        x=df4['relative_time [ms]'],
        y=df4['Bfield [microteslas]'],
        mode='markers',
        marker=dict(color='green', size=2),
        showlegend=True
    ),
])

fig.show()
```

Code used for polinomial fit is from google collab make-a-plot-and-do-a-fit.ipynb access by the link https://colab.research.google.com/drive/1BCUn4-EDf3bOB3jOjb6ryjxT7MKgkwDz?usp=sharing

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('https://raw.githubusercontent.com/pbelousova/471-Lab/main/B1%20-%20Sheet1%20(1).csv',delimiter=',',skiprows=1)
times = data[:,0]
distance = data[:,1]
plt.plot(times,distance,'b+',label='data')
```

```python
coefficients = np.polyfit(times,distance, 7)
print(coefficients)
poly = np.poly1d(coefficients)
fitX = np.linspace(data[0,0],data[-1,0],100)
fitY = poly(fitX)
plt.plot(times,distance,'b+',label='Raw data')
plt.plot(fitX,fitY,'r',label='nth order Poly Fit')
plt.legend()
plt.xlabel('Time [s]')
plt.ylabel('Distance [m]')
plt.title('nth order fit to our data')
plt.grid(True)
plt.show()
```
